```python
import numpy as np
import pandas as pd
from pathlib import Path
import yaml
import os 
```

```python
# Create output directory
folder = Path("/Users/paulianoprescu/Projects/wofost_miscanthus/miscanthus_calibration")
output_dir = folder / "input_params" / "agromanagement"
```

```python
# Define Agromanagement Assumptions
# Hereby, this notebook is reproduceable, assumptions are well documented, and if there is a change needed, everything can update accordingly
#
# Since Miscanthus is a perennial, each growing season is modelled as a different campaign 

crop_name = "miscanthus"
variety_name = "miscanthus sinensis"
crop_start_type = "emergence"   # rhizome sprouting = emergence; have not chosen for sowing
crop_end_type = "maturity"  # maturity is modelled for in this each growing season campaign, since harvest essentially happens just before emergence
max_duration = 300

# Growing Season Window (month, day)
start_month, start_day = 4, 1   # My observation in 2026 was that harvest happened in late March / early April; hereby I assume this emergance date 
end_month, end_day = 11, 15    # At this point in time I assume the crop to be fully grown

# Years to simulate (Planted in 2013; 2013-2014 establishment period; 2015-2024 observations)
years = list(range(2015, 2025))

# Create Crop Calendar File
campaigns = "\n".join([f"""- {year}-01-01:
    CropCalendar:
        crop_name: {crop_name}
        variety_name: {variety_name}
        crop_start_date: {year}-{start_month:02d}-{start_day:02d}
        crop_start_type: {crop_start_type}
        crop_end_date: {year}-{end_month:02d}-{end_day:02d}
        crop_end_type: {crop_end_type}
        max_duration: {max_duration}
    TimedEvents: null
    StateEvents: null""" for year in years])

miscanthus_agro = f"""Version: 1.0.0
AgroManagement:
{campaigns}
- {years[-1] + 1}-01-01:
    CropCalendar: null
    TimedEvents: null
    StateEvents: null
"""

os.makedirs(output_dir, exist_ok=True)
outpath = output_dir / f"misc_agro.yaml"
with open(outpath, 'w') as f:
    f.write(miscanthus_agro)
print(f"Miscanthus AgroManagement CropCalender file written to: {outpath}")
```
